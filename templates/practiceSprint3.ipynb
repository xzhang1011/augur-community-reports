## Practice with Bokeh and Python

```python
from bokeh.io import output_file, show
from bokeh.layouts import row
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool

# makes the plots show in jupyter notebook
output_notebook()
```

```python
def make_plots():
    x = list(range(11))
    y0 = x
    y1 = [10 - i for i in x]
    y2 = [abs(i - 5) for i in x]

    # create three plots using figure()
    s1 = figure(title = "Hover Over Plot Points",
                plot_width=250, 
                plot_height=250, 
                background_fill_color="#fafafa")
    # draw coordinates as circles
    s1.circle(x, y0, size=12, color="#53777a", alpha=0.8)

    # using bokeh hover tool
    s1.add_tools(HoverTool(
    tooltips=[
        ("(x,y)", "(@x, @y)"),
    ],
        mode = 'vline'
    ))

    # put the results in a row and show what it looks like
    show(s1)
```

```python
make_plots()
```
