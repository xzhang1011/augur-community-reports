## Visualizations

```python
from bokeh.io import output_file, show
from bokeh.layouts import row, column
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool,Panel, Tabs
from bokeh.resources import INLINE
```

```python
#source = ColumnDataSource(data=df['date'])

# instantiating figure model
code_graph = figure(x_axis_type = "datetime", title = "Code Change Count/Week")

#name of x-axos
code_graph.xaxis.axis_label = 'Date'
   
# name of the y-axis
code_graph.yaxis.axis_label = 'Code Changes'
df['date'] = pd.to_datetime(df['date'])
x = df['date']
code_graph.line(x,df['code_change_count'], color='navy', alpha=0.5)
code_graph.circle(x,df['code_change_count'])


# graph 2
lines_graph = figure(x_axis_type = "datetime", title = "Lines of Code Added/Week")
lines_graph.xaxis.axis_label = 'Date'
lines_graph.yaxis.axis_label = 'Line Changes'

#df['date'] = pd.to_datetime(df['date'])
lines_graph.circle(df['date'],df['lines_code_count'], color='navy', alpha=0.5)

hover = HoverTool(
    tooltips = [('Code Changes', '@y'),
                ('Date', '@x{%F}')
               ],
    formatters={
        '@x' : 'datetime',
    },
    mode = 'vline')

code_graph.add_tools(hover)

tab1 = Panel(child=code_graph, title="Code Changes")
tab2 = Panel(child=lines_graph, title="line")

show(Tabs(tabs=[tab1, tab2]))
```
